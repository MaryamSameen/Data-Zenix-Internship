```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023.csv")

# Set style
sns.set(style="whitegrid")

# Create a wide figure for side-by-side plots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# 1. Poverty Headcount Ratio over Years
sns.lineplot(x='Year', y='Poverty Headcount Ratio (%)', data=df, ax=axs[0, 0], marker='o', color='crimson')
axs[0, 0].set_title("Poverty Rate Over Time")
axs[0, 0].set_ylabel("Poverty (%)")

# 2. Unemployment Rate over Years
sns.lineplot(x='Year', y='Unemployment Rate (%)', data=df, ax=axs[0, 1], marker='o', color='darkblue')
axs[0, 1].set_title("Unemployment Rate Over Time")
axs[0, 1].set_ylabel("Unemployment (%)")

# 3. GDP Growth over Years
sns.lineplot(x='Year', y='GDP Growth Rate (%)', data=df, ax=axs[0, 2], marker='o', color='green')
axs[0, 2].set_title("GDP Growth Rate Over Time")
axs[0, 2].set_ylabel("GDP Growth (%)")

# 4. Inflation Rate over Years
sns.lineplot(x='Year', y='Inflation Rate (%)', data=df, ax=axs[1, 0], marker='o', color='orange')
axs[1, 0].set_title("Inflation Rate Over Time")
axs[1, 0].set_ylabel("Inflation (%)")

# 5. Government Social Spending
sns.lineplot(x='Year', y='Government Social Spending (% of GDP)', data=df, ax=axs[1, 1], marker='o', color='purple')
axs[1, 1].set_title("Social Spending Over Time")
axs[1, 1].set_ylabel("Spending (% of GDP)")

# 6. External Debt
sns.lineplot(x='Year', y='External Debt (USD billions)', data=df, ax=axs[1, 2], marker='o', color='teal')
axs[1, 2].set_title("External Debt Over Time")
axs[1, 2].set_ylabel("Debt (Billion USD)")

# Adjust layout
plt.tight_layout()
plt.show()
```

## 📊 Poverty and Economic Indicator Analysis: Pakistan (2000–2023)

This analysis explores the trends and relationships among key socioeconomic indicators in Pakistan from 2000 to 2023. The dataset includes variables such as:

- **Population Growth**
- **GDP Growth Rate**
- **Inflation Rate**
- **Unemployment Rate**
- **Poverty Headcount Ratio**
- **Agriculture Growth Rate**
- **Government Social Spending (% of GDP)**
- **External Debt (USD billions)**
- **Climate Disasters (count)**

### 🔍 Key Insights:
- **Poverty** steadily declined from 2000 to around 2016, with slight fluctuations afterward.
- Higher **GDP Growth** and increased **Social Spending** appear to correlate with a reduction in poverty levels.
- **Inflation** and **Unemployment** show weaker direct correlations with poverty but may have indirect effects.
- Although infrequent, **Climate Disasters** potentially impact both poverty and economic growth.

This exploratory data analysis highlights the importance of stable economic growth and social safety nets in poverty alleviation efforts in Pakistan.
